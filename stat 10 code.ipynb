```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as pl
from scipy.stats import chisquare
import math
for k in (1,5,10,100):
    chi2tot = np.zeros(10000)
    for a in range(0,10000):
        x = np.random.normal(size = k)
        #normal distribution around 0 and sigma = 1.
        chi2 = 0
        for i in x: 
            #in this case
            chi2 += i*i
        chi2tot[a] = chi2
    pl.hist(chi2tot, density=True, bins = 'auto')
    pl.title("Probability Distribution of chi2, k = " + str(k))
    pl.show()
```
